```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("householdtask3.csv")
```

```python
display(data.head(10))
```

```python
plt.bar(data['year'],data['tot_hhs'])
plt.title("Comparision of Year vs Total no. of houses built")
plt.xlabel('year')
plt.ylabel('tot_hhs')
plt.show()
```

```python
plt.bar(data['year'],data['own_wm'])
plt.title("Owning washing machine with year")
plt.xlabel('year')
plt.ylabel('own_wm')
plt.show()
```

```python
plt.bar(data['own_wm_prop'],data['own'])
plt.title("People who own property vs people who own property and washing machine")
plt.xlabel('Owns Property and Washing Machine')
plt.ylabel('Owns Property')
plt.show()
```

```python
data['no_own_prop']=data['tot_hhs']-data['own']
data.head()
```

```python
s1=sum(data['no_own_prop'])
s2=sum(data['own_wm']-data['own_wm_prop'])
labels=['no owm prop','wm with no owm prop']
sizes =[s1,s2]
colors=['lightblue','coral']
plt.pie(sizes,labels =labels,colors=colors, autopct ='%1.1f%%',startangle=90)
plt.title('percentage of people who do not own any property but own washing machine')
plt.axis('equal')
plt.show()
```
